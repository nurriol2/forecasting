```python
def convert_epoch(k):
    return datetime.datetime.fromtimestamp(int(k)/1000.0).strftime("%Y/%m/%d")
```

```python
closeURL = "http://services.runescape.com/m=itemdb_oldschool/api/graph/7944.json"
close_page = requests.get(closeURL).json()["daily"]

fmt_close = {}
for k, v in close_page.items():
    fmt_close[convert_epoch(int(k))] = v
    
pprint.pprint(fmt_close)
```

```python
close_dates = list(fmt_close.keys())
earliest = close_dates[0]
recent = close_dates[-1]

print("EARLIEST {}".format(earliest))
print("RECENT {}".format(recent))
```

## Note & Observations

For most items (notably excluded are bonds), when you visit the official OSRS GX website, there are 2 interactive graphs. One showing what I call PRICE data and the other showing VOLUME data. 

PRICE groups the daily closing price (gp) and the daily average price (gp) datasets together under a single term. VOLUME is the number of units sold on a particular timestamp.

While using the 180 day viewer, *both* plots show concurrent data:  data for the current timestamp (the day the page is being viewed) and the previous 179 days. 

Progromatically, however, PRICE and VOLUME data is gathered differently. PRICE is requested from the OSRS GX API and VOLUME is webscraped directly from the official OSRS GX website. 

As a result, **the timestamps for PRICE and VOLUME datasets are fundamentally offset from one another**. 

It was my previous assumption that the data I could *see* on the website was exactly the data being retrieved. However, there are 2 major problems with this assumption.

1. The daily **closing** price of a tradeable could not possibly be known before the day is over. 

So, it is likely that the datapoint that is being viewed on the website (i.e. the interactive graph) is just the current price *since the last graph update*. This fact explains why the most recent PRICE timestamp requested from the API is 1 day before the dataset is being accessed; the request returns only static datapoints.

2. The traded volume for the current timestamp is dynamic information, until the day is over. 

As shown at the end of the program above, in 9 minutes since scraping the trade volume, the number of units moved changed. This behavior reflects the meaning of this feature:  after sufficiently long enough time, buy/sell offers will be made/fufilled, thus changing the number of units moved. 


## Moving forward

After realizing these two flaws in my assumptions about the data, I created a short term solution and set a long term goal.

### Short term solution

I will continue to use the data collected directly from Jagex until a minimum viable product is achieved (aka 80% rule).

This choice means that the core time series (PRICE and VOLUME) will have at most 178 data points. Under the heuristic of "your (machine learning) model is only as good as your data", I think doing so is the best choice with respect to accuracy. 

In terms of the amount of data to train on, that is where the long term goal begins.

### Long term goal - Data collection

Third-party GX sites like GE-Tracker, OSRS Wiki, and RSBuddy have larger datasets on most tradeables. Especially when considering the trade volume datasets, these sources cannot be ignored.

Furthermore, I fully expect many ML models to perform inherently better with a greater amount of data. This expectation is another reason to consider collecting data from third-party sources.

However, the requirements for collecting this data would only further postpone MVP production. At the outset, I think it would be necessary to verify these sources with the dataset from official Jagex sources instead of blindly trusting them. On GE tracker alone, there are notable differences in the daily average price of chinchompas. 

It is more important to me that the "outline" of the project be finished before working any one part too deeply; basically, feeding as much energy to reaching the 80% rule as possible.
